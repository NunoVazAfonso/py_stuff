```python
# import and verify libs
import cv2
cv2.__version__
import numpy as np
import matplotlib.pyplot as plt

# to plot images conveniently
%matplotlib inline

# read input image
img = cv2.imread('/home/nunovazafonso/Downloads/burano.jpg')
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Convert image to Gray Scale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray, cmap = 'gray')
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))

for i in range(0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:, :, i], cmap = 'gray')
plt.show()
```

```python
# Transform the image into HSV and HLS models
# Hue, Saturation and Value or Lightness

img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)

# Plot the converted images
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize = (20, 20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

```python
# Step 1. Define callback function
def draw_circle(event, x, y, flags, param):    
    
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, center = (x, y), radius = 5, 
                   color = (87, 184, 237), thickness = -1)    
            
    elif event == cv2.EVENT_RBUTTONDOWN:        
        cv2.circle(img, center = (x, y), radius = 10,  
                   color = (87, 184, 237), thickness = 1)
            
```

```python
# Step 2. Call the window
img = cv2.imread('/home/nunovazafonso/Downloads/map.jpg')
cv2.namedWindow(winname = 'my_drawing')
cv2.setMouseCallback('my_drawing', draw_circle)
```

```python
while True:
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(10) & 0xFF == 27:
        breakcv2.destroyAllWindows()
        
```

```python
# Initialization
drawing = False
ix = -1
iy = -1

# create a drawing function
def draw_rectangle(event, x, y, flags, params):
    
    global ix, iy, drawing    
    
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y
        
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img, pt1=(ix, iy), pt2=(x, y), 
                          color = (87, 184, 237), thickness = -1)
            
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, pt1=(ix, iy), pt2=(x, y), 
                     color = (87, 184, 237), thickness = -1)
```
